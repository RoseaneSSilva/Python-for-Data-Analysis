1 Extração de coluna de arquivo csv

Extraia os valores valor_venda e armazena em uma lista.

```python
valores_vendas = []

with open (file='./carros.csv', mode='r', encoding='utf8') as arq:
  linha = arq.readline()
  linha = arq.readline()
  while linha:
    linha_separada = linha.split(sep = ',')

    valor_venda = linha_separada[1]
    valores_vendas.append(valor_venda)
    linha = arq.readline()

print(valores_vendas)
#print(len(valores_vendas))
```

Complete a função abaixo para extrair uma coluna, do arquivo csv em uma lista.

```python
def extrai_coluna_csv(nome_arquivo: str, indice_coluna: int):
    coluna = []

    with open (file=nome_arquivo, mode='r', encoding='utf8') as arq:
      linha = arq.readline()
      linha = arq.readline()
      while linha:
        linha_separada = linha.split(sep=',')
        colunas = linha_separada[indice_coluna]
        colunas = str(colunas)
        coluna.append(colunas)
        linha = arq.readline()
    return (coluna)
```

1. Funções para arquivo csv

Complete a função abaixo para extrair uma coluna do arquivo csv em uma lista. Os elementos devem ter o tipo de dado correto.

```python
def extrai_coluna_csv(nome_arquivo: str, indice_coluna: int, tipo_dado: str):

  coluna = []

  with open(file=nome_arquivo, mode='r', encoding='utf8') as arq:
    linha = arq.readline()
    linha = arq.readline()
    while linha:
      linha_separada = linha.split(sep=',')
      colunas = linha_separada[indice_coluna]
      if type(colunas) == type(tipo_dado):
        coluna.append(colunas)
      else:
          colunas = type(tipo_dado(colunas))
          coluna.append(colunas)
      linha = arq.readline()
      
  return coluna
```

```python
valor_venda = extrai_coluna_csv(nome_arquivo='./carros.csv', indice_coluna=1, tipo_dado='str')
print(valor_venda) 
```

```python
pessoas = extrai_coluna_csv(nome_arquivo='./carros.csv', indice_coluna=4, tipo_dado='int')
print(pessoas)
```
